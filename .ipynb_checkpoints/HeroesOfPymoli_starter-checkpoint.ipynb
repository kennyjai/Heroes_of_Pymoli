### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#Display the total number of players in dataframe
total_players = pd.DataFrame([{"Total Players": len(purchase_data["SN"].unique())}])
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate number of unique items
unique_items = len(purchase_data["Item ID"].unique())

# Calculate average price
average_price = "${:.2f}".format(purchase_data["Price"].mean())

# Calculate number of purchases
number_purchases = len(purchase_data["Purchase ID"].unique())

# Calculate total revenue
total_revenue = "${:.2f}".format(purchase_data["Price"].sum())

# Display summary of purchasing analysis
purchasing_analysis = pd.DataFrame([{
    "Number of Unique Items": unique_items,
    "Average Price": average_price,
    "Number of Purchases": number_purchases,
    "Total Revenue": total_revenue}])

# organize columns to better display data
purchasing_analysis[["Number of Unique Items","Average Price","Number of Purchases","Total Revenue"]]
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#summary of genders
gender_df = pd.DataFrame(purchase_data["Gender"].value_counts())
gender_demo = gender_df.rename(columns = {"Gender":"Total Count"})

# Add Percentage of Players
percentage_gender = (gender_demo["Total Count"] / len(purchase_data["SN"].unique())) * 100
gender_demo["Percentage of Players"] = percentage_gender

# Apply formatting to the table
pd.options.display.float_format = '{:.2f}'.format

# Sort in order we want
gender_demo[["Percentage of Players","Total Count"]]
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# group by gender
group_by_gender = purchase_data.groupby(["Gender"])

# Calculate purchase count
gender_pc = group_by_gender["Purchase ID"].count()

# Calculate average purchase price
gender_pp = group_by_gender["Price"].mean()

# Calculate total purchase value
gender_pv = group_by_gender["Price"].sum()

# Calculate normalized totals
gender_nt = gender_pv / gender_pc

# Summary of group by gender data
group_gender_summary = pd.DataFrame({"Purchase Count": gender_pc,
                                     "Average Purchase Price": gender_pp,
                                     "Total Purchase Value": gender_pv,
                                     "Normalized Totals": gender_nt})
pd.options.display.float_format = '${:,.2f}'.format
group_gender_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# using bins, create a new column that labels age into appropriate bins
df = pd.DataFrame(purchase_data)
df["Age Group"] = pd.cut(df["Age"],age_bins,labels=group_names)

# Create new dataframe to get total Count 
age_df = pd.DataFrame(df["Age Group"].value_counts())
age_df = age_df.rename(columns={"Age Group": "Total Count"})

# Calculate Percentage of Players
age_percentage = df["Age Group"].value_counts() / len(purchase_data["SN"].unique()) * 100
age_df["Percentage of Players"] = age_percentage
age_df = age_df[["Percentage of Players","Total Count"]]

# Apply formatting to the table
pd.options.display.float_format = '{:.2f}'.format

# sort by index
age_df.sort_index(inplace=True)
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create new df to calculate analysis based off age group
group_by_age = df.groupby("Age Group")

# Calculate purchase count
age_pc = group_by_age["Purchase ID"].count()

# Calculate average purchase price
age_pp = group_by_age["Price"].mean()

# Calculate total purchase value
age_tp = group_by_age["Price"].sum()

# Calculate normalized totals
age_nt = age_tp / age_pc

# Summary of group by gender data
group_age_summary = pd.DataFrame({"Purchase Count": age_pc,
                                     "Average Purchase Price": age_pp,
                                     "Total Purchase Value": age_tp,
                                     "Normalized Totals": age_nt})
pd.options.display.float_format = '${:,.2f}'.format
group_age_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create new df to determine top spenders by grouping by SN
group_by_spender = df.groupby("SN")

# Calculate purchase count
spender_pc = group_by_spender["Purchase ID"].count()

# Calculate average purchase price
spender_pp = group_by_spender["Price"].mean()

# Calculate total purchase value
spender_tp = group_by_spender["Price"].sum()

# Summary of dataframe grouped by SN
top_spender_summary = pd.DataFrame({"Purchase Count": spender_pc,
                                     "Average Purchase Price": spender_pp,
                                     "Total Purchase Value": spender_tp})

# Formatting
pd.options.display.float_format = '${:,.2f}'.format

# sort by Total Purchase Value in decreasing order
top_five = top_spender_summary.sort_values("Total Purchase Value",ascending=False)

# display top five results
top_five.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a new dataframe using original data extracting only Item ID, Item Name,and Price
popular_df = purchase_data[["Item ID","Item Name","Price"]]

# Group by Item ID and Item Name
group_popular = popular_df.groupby(['Item ID','Item Name'])

# Calculate purchase count of each item
popular_pc = group_popular["Item ID"].count()

# Price of each Item
popular_ip = group_popular["Price"].mean()

# Calculate total purchase value
popular_tp = group_popular["Price"].sum()

# Create summary of most popular items
popular_items_summary = pd.DataFrame({"Purchase Count": popular_pc,
                                      "Item Price": popular_ip,
                                     "Total Purchase Value": popular_tp})

# Formatting
pd.options.display.float_format = '${:,.2f}'.format

#Sort the purchase count column in descending order
sorted_popular = popular_items_summary.sort_values("Purchase Count",ascending=False)
sorted_popular.head(5)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Use above table to sort by total purchase value in descending order
sorted_profitable = popular_items_summary.sort_values("Total Purchase Value",ascending=False)
sorted_profitable.head(5)
```
